```python
import pandas as pd
import numpy as mp 
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../kalas/Downloads/Dataset/CARS.csv')
```

```python
df.MSRP = df.MSRP.replace('[$,]','',regex=True).astype('int64')
```

```python
df.Invoice = df.Invoice .replace('[$,]','',regex=True).astype('int64')
```

```python
def price (x):
    if x > 10000 and x <=22000:
        return "Low_B"
    elif x>22000 and x<=45000:
        return "Mid_B"
    elif x>45000 and x <=90000:
        return "High_budget"
    else:
        return "Luxury"
```

```python
df['Category'] = df.MSRP.apply(price)
```

```python
print(df.Make.unique())
ma = input("enter the brand name :")
maa = df[df.Make == ma]

print(maa.Type.unique())
ty = input("enter the type :")
tyy = maa[maa.Type == ty]

print(tyy.Category.unique())
ca = input("enter the Category you want :")
cat = tyy[tyy.Category ==ca]

sb.barplot(x = cat.Model , y =cat.MSRP)
plt.xticks(rotation = 90)

plt.show()
```
